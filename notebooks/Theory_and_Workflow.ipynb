## Teslakit theory and workflow


The repository is divided in **three big sections**, the first one is the climate emulator and generation of synthetic time series of waves and water level conditions the ***stochastic climate emulator***, the second one is the ***hybrid downscaling*** to obtain runup estimation at the shore, and finally, the analysis of ***climate change*** scenarios and their influence on expected extreme events.


## The stochastic climate emulator - 01_Offshore ##

#### The stochastic climate emulator is based on Anderson et al. (2019)


It is built on the recognition that coastal conditions are the result of meteorological forcing, and that synoptic-scale meteorology is in turn a consequence of large-scale quasi-steady atmospheric and oceanic patterns (e.g., Hannachi & Legras, 1995)

Climatologists have widely accepted classification methodologies that decompose time-varying spatial datasets into a certain number of representative climatic states or patterns as flexible frameworks for assessing the nonlinear nature of asymmetrical oscillating phenomena (Huth et al., 2008; Philipp et al., 2016). A wide range of automated techniques exist (i.e., leader algorithms, optimized partitioning, etc.), and the resultant patterns can be dependent on the technique employed. A tailor-made climate index was developed for this study, largely due to a desire to preserve nonlinearity in the climate system and retain more information than off-the-shelf univariate indices provide (Williams & Patricola, 2018). The technique employed is a weather-type (WT)-based framework which has been used to explore climate variability in many of the physical processes relevant to coastal erosion and flooding extremes (e.g., Camus et al., 2014b). **The weather types in this work** are obtained through a two-step process of identifying dominate modes of spatial variability through empirical orthogonal functions (EOFs) and subsequent K-means clustering with respect to the temporal amplitude of each principal component (PCs) (Gutierrez et al., 2013). The resulting centroids for each cluster thus have an amplitude associated with each EOF and can be remapped to a spatial weather type considered to be representative of all instances associated with that centroid. Weather-typing was performed to identity **representative states of ENSO (AWT)** (***Notebook 01***), **MJO (IWT)** (***Notebook 04***) and representative **daily atmospheric circulation patterns (DWT)** (***Notebook 10***).


**Simulating sequencing and persistence of synthetic of AWT** (***Notebook 01***), **IWT** (***Notebook 04***) **and DWTs** (***Notebook 11***) is accomplished with **autoregressive logistic models (ALR)**. ALR models are simultaneously able to account for covariates varying at different timescales as well as the autocorrelation of those covariates at different orders (Guanche et al., 2013; Antolinez et al., 2015). In this sense, the AWT, seasonality,intraseasonal variability(IWT), as well as the ordering (transitions between DWTs) and duration (persistence within a DWT) can all be accounted for within a single framework to make a categorical decision of what the weather pattern should be on any given day.

## Hybrid downscaling to the coast - 02_Nearshore##

Once continues validated time series are obtained offshore, they need to be transferred to the coast to assess the potential of flooding. The dynamical downscaling of these 1000s of years of multivariate events is unaffordable, therefore an ***hybrid method*** has been developed. The method is similar to the one used to downscaled TC waves, where a combination of statistical tools (MDA ***Notebook 1***  and RBFs ***Notebook 3***) and numerical models (in this case the nearshore wave model SWAN ***Notebook 2***) are used to downscale the waves from offshore to the nearshore. And to estimate inundation extents, run-up (R2%) estimations are calculated with HyCReWW metamodel ***Notebook 4***
